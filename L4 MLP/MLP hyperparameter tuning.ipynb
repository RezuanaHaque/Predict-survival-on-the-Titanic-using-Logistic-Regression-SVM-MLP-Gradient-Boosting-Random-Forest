```python
import joblib
import pandas as pd
from sklearn.model_selection import GridSearchCV
# we're going to use this warnings package to suppress these warnings
import warnings
warnings.filterwarnings('ignore', category = FutureWarning)
warnings.filterwarnings('ignore', category = DeprecationWarning)
tr_features= pd.read_csv('/ML/Applied ML Algorithms/L1/train_features.csv')
tr_labels=pd.read_csv('/ML/Applied ML Algorithms/L1/train_label.csv')
tr_labels.head(5)
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, results.cv_results_['params']):
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
mlp = MLPClassifier()
parameters = {
    'hidden_layer_sizes': [(10,), (50,), (100,)],
#      the second entry in this tuple would represent the number of layers leaving it blank, just implies one.
    'activation': ['relu', 'tanh', 'logistic'],
    'learning_rate': ['constant', 'invscaling', 'adaptive']
}
cv = GridSearchCV(mlp, parameters, cv=5)
cv.fit(tr_features, tr_labels.values.ravel())

print_results(cv)
```

the first is constant. So that'll just take that initial learning rate and it'll keep the same throughout the entire optimization process. The second is invscaling in other words, inverse scaling. So it gradually decreases the learning rate at each step. So this will allow it to take large jumps at first, and then it slowly decreases as it gets closer and closer to the optimal model. And the last is adaptive and this keeps the learning rate constant, as long as the training loss keeps decreasing. If the learning rate stops going down, then it will decrease the learning rate so that it takes smaller steps

we have three settings for hidden layer sizes, three settings for activation and three settings for learning rate. So altogether that's 27 separate models that is being fit within GridSearchCV, and every time it's being fit, it's throwing this model. So essentially we see it 27 times.
